### Déclaration des fonctions

```python
# Déclaration des fonctions correlation et tauxDeRemplissage
def correlation(dataframe, method):
  liste = []
  corr = dataframe.corr(method = method)
  colonnesIndex = corr.index
  colonnesNoms = corr.columns
  for colonne in colonnesIndex:
    for colonne2 in colonnesNoms:
      if colonne != colonne2:
        trie = [colonne, colonne2]
        trie.sort()
        liste.append([trie[0], trie[1], corr.loc[colonne, colonne2]])
  return pd.DataFrame(liste, columns=["col1", "col2", "values"]).drop_duplicates(["col1", "col2"])

def tauxDeRemplissage(df):
  return (df.notna().sum()*100/len(df)).sort_values()
```

## 5- Test anova
*Compare une feature numérique rapport a des features catégorielles ou des features numériques entre elles*

```python
# Réalisation d'un Anova
df_anova = df.copy()
df_anova['Satisfaction'] = df_anova['Satisfaction'].replace({'satisfied': 1, 'neutral or dissatisfied': 0})

#Gender + Customer_Type + Age + Type_of_Travel + Class + Flight_Distance + Seat_comfort + Departure_Arrival_time_convenient + 
#Food_and_drink + Gate_location + Inflight_wifi_service + Inflight_entertainment + Online_support + Ease_of_Online_booking + 
#On_board_service + Leg_room_service + Baggage_handling + Checkin_service + Cleanliness + Online_boarding + Departure_Delay_in_Minutes + Arrival_Delay_in_Minutes'

model = ols('Satisfaction ~ Gender + Customer_Type + Age + Type_of_Travel + Class + Flight_Distance + Seat_comfort + Departure_Arrival_time_convenient + Food_and_drink + Gate_location + Inflight_wifi_service + Inflight_entertainment + Online_support + Ease_of_Online_booking + On_board_service + Leg_room_service + Baggage_handling + Checkin_service + Cleanliness + Online_boarding + Departure_Delay_in_Minutes + Arrival_Delay_in_Minutes', data=df_anova)
entrainement = model.fit()
anova_lm(entrainement)
```

Conclusion
- l'anova nous permet de comparer une feature numériques avec des features catégorielles
- pour l'exploiter il faut que PR(>F) soit inférieur à 5% (0.05) pour que la donnée explique la feature numérique de référence
- à l'inverse si la données est supérieur à 5% (0.05) elle peux être écartée de la justification
- ici la plupart des features sont proche de zéro donc elles participent à l'explication de la Satisfaction
- en revanche la feature age qui est à 0.24 soit 24% peut être écartée des features qui expliquerais la satisfaction

## 6- Test anova avec map()

```python
# Réalisation du test Anova
df_anova = df.copy()
df_anova['Satisfaction'] = df_anova['Satisfaction'].map({'satisfied': 1, 'neutral or dissatisfied': 0}).astype(int)

#Gender + Customer_Type + Age + Type_of_Travel + Class + Flight_Distance + Seat_comfort + Departure_Arrival_time_convenient + 
#Food_and_drink + Gate_location + Inflight_wifi_service + Inflight_entertainment + Online_support + Ease_of_Online_booking + 
#On_board_service + Leg_room_service + Baggage_handling + Checkin_service + Cleanliness + Online_boarding + Departure_Delay_in_Minutes + Arrival_Delay_in_Minutes'

model = ols('Satisfaction ~ Gender + Customer_Type + Age + Type_of_Travel + Class + Flight_Distance + Seat_comfort + Departure_Arrival_time_convenient + Food_and_drink + Gate_location + Inflight_wifi_service + Inflight_entertainment + Online_support + Ease_of_Online_booking + On_board_service + Leg_room_service + Baggage_handling + Checkin_service + Cleanliness + Online_boarding + Departure_Delay_in_Minutes + Arrival_Delay_in_Minutes', data=df_anova)
entrainement = model.fit()
anova_lm(entrainement)
```

## 7- Test anova avec encodage

```python
# Test anova avec des données encodées
df_anova_encod = df.copy()

# Utilisation de LabelEncoder()
le = LabelEncoder()
df_anova_encod['Satisfaction'] = le.fit_transform(df_anova_encod['Satisfaction'])

# Utilisation de get_dummies()
data_encoded = pd.get_dummies(df_anova_encod, columns=['Gender', 'Class', 'Type_of_Travel', 'Customer_Type'])

# Initialisation du DataFrame pour stocker les résultats de l'ANOVA
results_df = pd.DataFrame(columns=['feature', 'F-value', 'p-value'])

# Calcul de l'ANOVA pour chaque colonne
for col in data_encoded.columns:
    groups = []
    for val in data_encoded[col].unique():
        group = df_anova_encod[data_encoded[col] == val]['Satisfaction']
        groups.append(group)
    f_stat, p_value = f_oneway(*groups)
    results_df.loc[len(results_df)] = [col, f_stat, p_value]

results_df
```

Conclusion
- quand on vois le p-value, l'age n'est plus a écarté comparer aux autres tests anova

## 8- Test chi2
*Permet d'expliquer des features catégorielles entre elles*

```python
# Récupération d'un DataFrame pour le test chi2
chi2 = df.copy()
```

### Chi2 feature par feature

```python
# Test chi2 feature par feature

# récupération d'une liste des features catégorielles
colonnes_cat_chi2 = list(chi2.select_dtypes(include=['object']).columns)
colonnes_cat_chi2.remove('Satisfaction')

# déclaration d'un dataframe pour accueillir le résultat
results_df_chi2 = pd.DataFrame(columns=['feature', 'p-value'])

# calcul des chi2 et mise a jour du dataframe résultat
for i in colonnes_cat_chi2:
    st_chi2, st_p, st_dof, st_exp  = chi2_contingency(chi2[["Satisfaction", i]].pivot_table(index="Satisfaction", columns=i, aggfunc=len))

    results_df_chi2.loc[len(results_df_chi2)] = [i, st_p] 

results_df_chi2
```

### Chi2 entre satisfaction et des combinaisons de features

```python
# Test chi2 avec toute les features
# Création d'une table de contingence
table = pd.crosstab(chi2["Satisfaction"], [chi2["Gender"], chi2["Class"], chi2["Type_of_Travel"], chi2["Customer_Type"]], margins=True, margins_name='Total')

# Effectuez un test du Chi-carré en utilisant la bibliothèque SciPy
chi2, p_value, dof, expected = chi2_contingency(table)

'''# Affichage des résultats
print('Résultats du test du Chi-carré')
print('==============================')
print('Chi2 :', chi2)
print('Degrés de liberté :', dof)
print('p-value :', p_value)
print('Fréquences attendues : \n')'''

df_result = pd.DataFrame(expected)
df_result
```

```python
# Calcul de la différence entre les fréquences observées et attendues
diff = table - expected
diff.drop(columns=['Total'], inplace=True)
diff.drop(["Total"], inplace=True)
diff = diff.transpose()


# Affichage du résultat sous forme de heatmap
sns.heatmap(diff, cmap="coolwarm", fmt=".2f")
```

### Conclusion
- Avec le test chi on peux voir que indépendamment les features catégorielles influent toutes sur la satisfaction. On peux aussi voir avec la dernière heatmap que plus précisement, certaines combinaisons des features categorielles ont beaucoup d'influence sur la satisfaction ou la non satisfaction.

## Conclusion générale

Dans cette analyse j'ai essayé plusieurs façons de trouver des corrélations entre les données :
- **les matrices de correlation :** qui permettent de trouver des corrélations entre les features numérique. Le problème dans le projet est que notre target est une feature catégorielle et qu'il y a dans la table plusieurs autres données catégorielles. Ceci impose le traitement des données avant de créer des matrices de corrélation et ne nous assure pas une interprétation fidèle à la réalité.
- **le test anova :** le test anova permet de trouver des corrélations entre des données numériques ou une donnée catégorielle et des données numérique. Satisfaction étant notre target, une erreur m'a obligé à la passé en donnée numérique. J'ai essayé de transformer la target de plusieurs façon, simplement avec replace(), map() ou un encodage, les résultats de ces différents éssais ne varient pas trop. Toute les features sauf éventuellement la feature Age, expliquent la feature satisfaction.
- **le test chi2 :** le chi2 permet de trouver des corrélations entre des données catégorielles. J'ai comparer Satisfaction a chacune des autres données catégorielles indépendamment et il en ressort qu'elles expliquent toutes la donnée Satisfaction. Pour préciser j'ai comparer Satisfaction à l'ensemble des combinaisons des données catégorielles ce qui met en avant certaines combinaisons.

- Female - Business - Business Travel - Loyal Customer
- Female - Business - Business Travel - disloyal Customer
- Female - Eco      - Business Travel - disloyal Customer
- Female - Eco      - Personal Travel - Loyal Customer

- Male - Business - Business Travel - Loyal Customer
- Male - Eco      - Business Travel - disloyal Customer
- Male - Eco      - Personal Travel - Loyal Customer
- Male - Eco Plus - Personal Travel - Loyal Customer

Les features qui semblent être assez dominantes sont Inflight_entertainment (divertissement en vol), Ease_of_Online_booking (la facilité de réservation), Online_support	On_board_service, ce qui est lié au service ce qui n'est pas abérrant et l'age n'entre pas en ligne de compte.
